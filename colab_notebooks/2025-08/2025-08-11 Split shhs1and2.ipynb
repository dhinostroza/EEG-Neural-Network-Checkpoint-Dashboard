```python
python
import shutil
from pathlib import Path
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

print("--- INICIANDO LA ORGANIZACIÓN DE ARCHIVOS .parquet ---")

source_dir = Path('/content/drive/MyDrive/shhs_processed/')
destination_dir = Path('/content/drive/MyDrive/shhs2_processed/')

destination_dir.mkdir(parents=True, exist_ok=True)
print(f"Directorio de destino asegurado en: {destination_dir}")

files_to_move = list(source_dir.glob('shhs2*.parquet'))

if not files_to_move:
    print("No se encontraron archivos de shhs2 para mover. Puede que ya estén organizados.")
else:
    print(f"Se encontraron {len(files_to_move)} archivos de shhs2 para mover.")

    for file_path in files_to_move:
        try:
            shutil.move(str(file_path), str(destination_dir))
        except Exception as e:
            logging.error(f"No se pudo mover {file_path.name}. Razón: {e}")

    print(f"✅ Proceso de movimiento completado. {len(files_to_move)} archivos movidos a {destination_dir}.")

print("--- ORGANIZACIÓN DE ARCHIVOS COMPLETADA ---")
```
